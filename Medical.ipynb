```python
import streamlit as st
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model
from transformers import pipeline, BertTokenizer, BertForSequenceClassification
from PIL import Image

# Load Pre-trained CNN Model for Medical Imaging
@st.cache_resource
def load_cnn_model():
    base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
    x = Flatten()(base_model.output)
    x = Dense(128, activation='relu')(x)
    output = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=base_model.input, outputs=output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Load Pre-trained NLP Model for Text Summarization
@st.cache_resource
def load_nlp_model():
    model_name = "bert-base-uncased"
    tokenizer = BertTokenizer.from_pretrained(model_name)
    model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)
    nlp_pipeline = pipeline("text-classification", model=model, tokenizer=tokenizer)
    return nlp_pipeline

# Function to Predict Image Diagnosis
def predict_image(model, img_array):
    prediction = model.predict(img_array)
    return "Tumor Detected" if prediction[0] > 0.5 else "No Tumor Detected"

# Function to Summarize Patient History
def summarize_history(nlp_pipeline, text):
    result = nlp_pipeline(text)
    return result[0]['label']

# Streamlit App Interface
st.title("AI-Powered Medical Diagnosis System")

# Sidebar Menu
st.sidebar.title("Menu")
options = st.sidebar.radio("Choose an option", ["Image Diagnosis", "Patient History", "Chatbot"])

# Load Models
cnn_model = load_cnn_model()
nlp_model = load_nlp_model()

if options == "Image Diagnosis":
    st.header("Upload Medical Image")
    uploaded_file = st.file_uploader("Upload an X-ray, MRI, or CT scan", type=["jpg", "jpeg", "png"])

    if uploaded_file:
        img = Image.open(uploaded_file).resize((224, 224))
        st.image(img, caption="Uploaded Image", use_column_width=True)
        img_array = np.array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        # Predict using the CNN model
        result = predict_image(cnn_model, img_array)
        st.write(f"Diagnosis: **{result}**")

elif options == "Patient History":
    st.header("Summarize Patient History")
    user_text = st.text_area("Enter patient history:")

    if st.button("Summarize"):
        summary = summarize_history(nlp_model, user_text)
        st.write(f"Summary: **{summary}**")

elif options == "Chatbot":
    st.header("Doctor-Patient Chatbot")
    chatbot = pipeline("text-generation", model="gpt2")

    user_query = st.text_input("Ask your question:")
    if user_query:
        response = chatbot(user_query, max_length=50)
        st.write(f"Response: {response[0]['generated_text']}")
```
